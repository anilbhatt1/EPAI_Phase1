**Image crop - Pixel Based**

**Fits the image size**

im_crop = im.crop((left, upper, right, lower))

(left, upper, right, lower) means two points

(left, upper)
(right, lower)
with an 800x600 pixel image, the image's left upper point is (0, 0), the right lower point is (800, 600).

```python
img = plt.imread(jpg_path)
plt.imshow(img)
plt.show()
```

```python
im = Image.open(jpg_path)
crp_px = (20, 25, 200, 100)

left, upper, right, lower = crp_px
crop_height = lower - upper
crop_width = right - left
width, height = im.size
if crop_width > width or crop_height > height:
    print('Unable to crop : Crop dimensions > Image dimensions')
else:
    img_crop = im.crop(crp_px)
    img_crop.save(cropped_jpg_path)
    im1 = Image.open(cropped_jpg_path)
    np_arr = np.array(im1)
    np_arr.shape
    print(crp_px, crop_height, crop_width, width, height, np_arr.shape )
```

```python
img = plt.imread(cropped_jpg_path)
plt.imshow(img)
plt.show()
```

```python
im = Image.open(png_path)
crp_p = 0.5
width, height = im.size
remove = 1 - crp_p
left = int(width* remove/2)
right = int(left + width*crp_p)
upper = int(height*remove/2)
lower = int(upper+height*crp_p)
crp_px = (left,upper,right,lower)

if right > width or lower > height:
    print('Unable to crop : Crop dimensions > Image dimensions')
else:
    img_crop = im.crop(crp_px)
    img_crop.save(cropped_png_path)
    im1 = Image.open(cropped_png_path)
    np_arr = np.array(im1)
    np_arr.shape
    print(crp_p, crp_px, width, height, np_arr.shape )
```
